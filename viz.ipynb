```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
data_df = pd.read_csv("tournamentdata/new/playerstats_cleaned1.csv")
data_df.head()
```

```python
x_axis = np.arange(len(data_df))
tick_locations = [value for value in x_axis]
```

```python
grouped_df = data_df.sort_values(
    by="Kills",
    ascending=False)
grouped_df
```

```python
teams_df = grouped_df.sort_values(
    by="Kills",
    ascending=False)
teams_df
```

```python
plt.figure(figsize=(14,10))
plt.bar(x_axis, grouped_df["Kills"], color='r', alpha=0.5)
plt.xticks(tick_locations, grouped_df["Player"], rotation="vertical")
plt.title("Total Amount of Kills (Ordered)")
plt.xlabel("Name of Player")
plt.ylabel("Number of Kills")
plt.savefig("visuals/KillsOrdered.png")
plt.show()
```

```python
grouped_df = data_df.sort_values(
    by="Deaths",
    ascending=False)
grouped_df
```

```python
plt.figure(figsize=(14,10))
plt.bar(x_axis, grouped_df["Deaths"], color='brown', alpha=0.5)
plt.xticks(tick_locations, grouped_df["Player"], rotation="vertical")
plt.title("Total Amount of Deaths (Ordered)")
plt.xlabel("Name of Player")
plt.ylabel("Number of Deaths")
plt.savefig("visuals/DeathsOrdered.png")
plt.show()
```

```python
grouped_df = data_df.sort_values(
    by="Assists",
    ascending=False)
grouped_df
```

```python
plt.figure(figsize=(14,10))
plt.bar(x_axis, grouped_df["Assists"], color='g', alpha=0.5)
plt.xticks(tick_locations, grouped_df["Player"], rotation="vertical")
plt.title("Total Amount of Assists (Ordered)")
plt.xlabel("Name of Player")
plt.ylabel("Number of Assists")
plt.savefig("visuals/AssistsOrdered.png")
plt.show()
```

```python
grouped_df = data_df.sort_values(
    by="ACS/Map",
    ascending=False)
grouped_df
```

```python
# Average Combat Score is the cumulation of Kills, Damage and Non-Damaging Assists in a map.
plt.figure(figsize=(14,10))
plt.bar(x_axis, grouped_df["ACS/Map"], color='y', alpha=0.5)
plt.xticks(tick_locations, grouped_df["Player"], rotation="vertical")
plt.title("Best ACS (Ordered)")
plt.xlabel("Name of Player")
plt.ylabel("ACS")
plt.savefig("visuals/ACSOrdered.png")
plt.show()
```

```python
grouped_df = data_df.sort_values(
    by="KDA",
    ascending=False)
grouped_df
```

```python
reducedteams_df = teams_df[["Player", "Team", "Kills", "Deaths"]]
reducedteams_df.head()
```

```python
reducedteams_df["Team Player"] = reducedteams_df["Team"] + ' ' + reducedteams_df['Player']
print(reducedteams_df)
```

```python
reducedteams_df.head()
```

```python
# Removed "Player" and "Team" as they were combined. 
reducedteams_df = reducedteams_df.drop(["Player", "Team"], axis = 1)
print(reducedteams_df)
```

```python
sen_team_stat = reducedteams_df[reducedteams_df["Team Player"].str.contains("Sentinels")][['Kills', "Deaths"]].sum()
print(sen_team_stat)
```

```python
gen_team_stat = reducedteams_df[reducedteams_df["Team Player"].str.contains("Gen.G Esports")][['Kills', "Deaths"]].sum()
print(gen_team_stat)
```

```python
multi_plot = reducedteams_df.plot(kind="bar", figsize=(20,20))

multi_plot.set_xticklabels(reducedteams_df["Team Player"], rotation=90)

bar_width = 0.35
plt.bar(reducedteams_df['Kills'], bar_width, label='Kills', color='r')
plt.bar(reducedteams_df['Deaths'], bar_width, label='Deaths', color='b')

plt.title("Kill / Death Ratio of Each Player")
plt.xlabel("Team and Name of Player")
plt.ylabel("Number of Kills and Deaths")
plt.savefig("visuals/KDeveryplayer.png")
plt.show()
```
