```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
data_df = pd.read_csv("tournamentdata/new/playerstats_cleaned1.csv")
data_df.head()
```

```python
x_axis = np.arange(len(data_df))
tick_locations = [value for value in x_axis]
```

```python
plt.figure(figsize=(16,4))
plt.bar(x_axis, data_df["Kills"], color='r', alpha=0.5)
plt.xticks(tick_locations, data_df["Team"], rotation="vertical")
```

```python
grouped_df = data_df.sort_values(
    by="Kills",
    ascending=False)
grouped_df
```

```python
plt.figure(figsize=(16,4))
plt.bar(x_axis, grouped_df["Kills"], color='r', alpha=0.5)
plt.xticks(tick_locations, grouped_df["Player"], rotation="vertical")
```

```python
teams_df = grouped_df.sort_values(
    by="Kills",
    ascending=False)
teams_df
```

```python
reducedteams_df = teams_df[["Player", "Team", "Kills", "Deaths", "Assists"]]
reducedteams_df.head()
```

```python
multi_plot = reducedteams_df.plot(kind="bar", figsize=(20,5))

# PandasPlot.set_xticklabels() can be used to set the tick labels as well
multi_plot.set_xticklabels(reducedteams_df["Player"], rotation=45)

plt.show()
plt.tight_layout()
```
